## STEP 1

- Went on rapidapi.com and searched for yhfinance -> [link-to-api](https://rapidapi.com/sparior/api/yahoo-finance15)

```python
import requests     # the python client we're using to interact with the api

url = "https://yahoo-finance15.p.rapidapi.com/api/v2/markets/tickers"   # base url + endpoint -> full URL

querystring = {"page":"2","type":"STOCKS"}  # define the query parameters 

# define the headers (auth) - headers just includes all necessary info to access the api (typically includes the api key)
headers = {
	"x-rapidapi-key": "",
	"x-rapidapi-host": "yahoo-finance15.p.rapidapi.com"
}

# Making the actual GET call:
response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

```python
url = "https://yahoo-finance15.p.rapidapi.com/api/v2/markets/tickers"  

headers = {
	"x-rapidapi-key": "",
	"x-rapidapi-host": "yahoo-finance15.p.rapidapi.com"
}

tickers = []	# empty list to store the symbols to

for page in range(1,2):
	querystring = {"page":str(page),"type":"STOCKS"}
	response = requests.get(url, headers=headers, params=querystring)	# contains the api response object from this GET request
	json_response = response.json()	 # in order to retrieve the data from the response object 'response' we must use .json()

	if 'body' in json_response and isinstance(json_response['body'], list):
		for stock in json_response['body']:
			current_ticker = stock.get('symbol')	# we are storing the current stock's symbol into a variable 'current_ticker'
			if current_ticker:	# I'm checking whether or not current_ticker actually contains something
				tickers.append(current_ticker)	# if current_ticker contains a ticker symbol (meaning the page contained ticker symbols and was not empty) then add it to the existing list 'tickers'.
```

```python
tickers = ['ORCL', 'KO', 'SAP', 'ABBV', 'TMUS']
```

```python
import requests
import time

news_url = "https://yahoo-finance15.p.rapidapi.com/api/v2/markets/news"

headers = {
	"x-rapidapi-key": "",
	"x-rapidapi-host": "yahoo-finance15.p.rapidapi.com"
}

news_dict = {}

for ticker in tickers:
    querystring = {"tickers":ticker,"type":"ALL"}
    response = requests.get(news_url, headers=headers, params=querystring)
    news_data = response.json()
    
    articles_list = []
    if 'body' in news_data and isinstance(news_data['body'], list):
        for article in news_data['body']:
            title = article.get('title')
            text = article.get('text')
            if title and text:
                articles_list.append([title, text])
                
    if articles_list:
            news_dict[ticker] = articles_list
    
    time.sleep(2)
```
